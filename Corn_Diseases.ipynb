```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
# Update the file path to the path in Google Drive
file_path = '/content/drive/My Drive/Execute file/knoledge base.txt'

class Rule:
    def __init__(self, facts, conclusion):
        self.facts = facts
        self.conclusion = conclusion


class InferenceForwardChaining:
    @staticmethod
    def do_forward_chaining(rules, initial_facts):
        inferred_facts = set(initial_facts)
        new_fact_added = True

        while new_fact_added:
            new_fact_added = False
            for rule in rules:
                # Check if all facts of the rule are in inferred_facts
                if set(rule.facts).issubset(inferred_facts):
                    if rule.conclusion not in inferred_facts:
                        inferred_facts.add(rule.conclusion)
                        new_fact_added = True

        return inferred_facts


class UI:
    def __init__(self):
        self.questions = []
        self.answers = [0] * 9  # Replace 5 with the actual number of symptoms if needed
        self.set_questions()

    def set_questions(self):
        self.questions.append("Does your leaves is Chlorotic Colored?")
        self.questions.append("Does your plant experiencing stunted growth?")
        self.questions.append("Do your leaves look wilted?")
        self.questions.append("re there several small spots that form a larger spot?")
        self.questions.append("Does your plant have elongated light brown spots in the form of a coil or boat?")
        self.questions.append("Also does it have a brown spots shaped like an ellipse?")
        self.questions.append("And lastly does the leaves look dry?")
        self.questions.append("Are there any leaves that look pretty dry?")
        self.questions.append("Are there any small brown or yellow spots on the leaf surface?")
        # Add more questions as per knowledge base

    def show_questions(self):
        for i, question in enumerate(self.questions):
            print(question)
            answer = input("0. No    1. Yes: ")
            while answer not in ['0', '1']:
                print("Answer correctly!")
                answer = input("0. No    1. Yes: ")
            self.answers[i] = int(answer)

    def get_facts(self):
        facts = set()
        if self.answers[0] == 1:
            facts.add("g1")
        if self.answers[1] == 1:
            facts.add("g2")
        if self.answers[2] == 1:
            facts.add("h1")
        if self.answers[3] == 1:
            facts.add("h2")
        if self.answers[4] == 1:
            facts.add("h3")
        if self.answers[5] == 1:
            facts.add("h4")
        if self.answers[6] == 1:
            facts.add("h5")
        if self.answers[7] == 1:
            facts.add("r1")
        if self.answers[8] == 1:
            facts.add("r2")
        return facts

    def show_conclusion(self, g_is_true, a2_is_true, a3_is_true, facts, inferred_facts):
        print("All Symptoms:", facts)
        print("Inferred Symptoms:", inferred_facts)
        if g_is_true:
            print("You plant got Disease: Bulai")
        if a2_is_true:
            print("You plant got: Blight")
        if a3_is_true:
            print("You plant got Disease: Leaf Sickness")


def get_knowledge():
    rules = []
    with open(file_path, 'r') as file:
        for line in file:
            # Skip empty lines
            if not line.strip():
                continue

            parts = line.strip().split('-')
            # Check if the line contains the delimiter
            if len(parts) < 2:
                print(f"Warning: Skipping line '{line.strip()}' due to missing delimiter.")
                continue  # Skip to the next line

            facts = parts[0].split(',')
            conclusion = parts[1]
            rules.append(Rule(facts, conclusion))
    return rules


def main():
    ui = UI()
    ui.show_questions()
    facts = ui.get_facts()

    # Get knowledge base
    rules = get_knowledge()

    # Apply forward chaining
    inferred_facts = InferenceForwardChaining.do_forward_chaining(rules, facts)

    g_is_true = 'a1' in inferred_facts
    a2_is_true = 'a2' in inferred_facts
    a3_is_true = 'a3' in inferred_facts
    ui.show_conclusion(g_is_true, a2_is_true, a3_is_true, facts, inferred_facts)


if __name__ == "__main__":
    main()
```
